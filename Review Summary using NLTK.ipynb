```python
# Removing Square Brackets and Extra Spaces
ShazamReview = re.sub(r'\[[0-9]*\]', ' ', ShazamReview)  
ShazamReview = re.sub(r'\s+', ' ', ShazamReview)
ShazamReview
```

```python
# Removing special characters and digits
formattedShazamReview = re.sub('[^a-zA-Z]', ' ', ShazamReview )  
formattedShazamReview = re.sub(r'\s+', ' ', formattedShazamReview)  
formattedShazamReview
```

```python
sentence_list = nltk.sent_tokenize(ShazamReview)
sentence_list
```

```python
stopwords = nltk.corpus.stopwords.words('english')

word_frequencies = {}  
for word in nltk.word_tokenize(formattedShazamReview):  
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
```

```python
word_frequencies
```

```python
maximum_frequncy = max(word_frequencies.values())

for word in word_frequencies.keys():  
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
```

```python
word_frequencies
```

```python
sentence_scores = {}  
for sent in sentence_list:  
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 30:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
```

```python
sentence_scores
```

```python
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

summary = ' '.join(summary_sentences)  
print(summary)  
```
